```
l1 = ['song','album','albumartist','artist','date','encodersettings','genre','title','duration']
l2 = []
i=0
import os
source = '/content/drive/My Drive/new dataset/SAD (Pranita)'
for root, dirs, filenames in os.walk(source):
    for file in filenames:
        print(file)
        metadata = audio_metadata.load(file)
        l2.append(file)
        print(l2)
        if hasattr(metadata.tags,'album'):
          item=str(metadata.tags.album)
          #print(item)
          it = item.strip("[']").split(" -")
          #print(it)
          l2.append(it[0])
          #print(l2)
        else:
          l2.append('null')

        if hasattr(metadata.tags,'albumartist'):
          a=str(metadata.tags.albumartist)
          #print(a)
          it = a.strip("[']").split(" -")
          #print(it)
          l2.append(it[0])
          #print(l2)
        else:
          l2.append('null')

        if hasattr(metadata.tags,'artist'):
          b=str(metadata.tags.artist)
          #print(b)
          it = b.strip("[']").split(" -")
          #print(it)
          l2.append(it[0])
          #print(l2)
        else:
          l2.append('null')

        if hasattr(metadata.tags,'date'):
          c=str(metadata.tags.date)
          #print(c)
          it = c.strip("[']")
          #print(it)
          l2.append(it)
          #print(l2)
        else:
          l2.append('null')

        if hasattr(metadata.tags,'encodersettings'):
          d=str(metadata.tags.encodersettings)
          print(d)
          it = d.strip("[']")
          #print(it)
          l2.append(it)
          #print(l2)
        else:
          l2.append('null')

        if hasattr(metadata.tags,'genre'):
          e=str(metadata.tags.genre)
          #print(e)
          it = e.strip("[']")
          #print(it)
          l2.append(it)
          #print(l2)
        else:
          l2.append('null')

        if hasattr(metadata.tags,'title'):
          f=str(metadata.tags.title)
          #print(f)
          it = f.strip("[']").split(" -")
          #print(it)
          l2.append(it[0])
          #print(l2)
        else:
          l2.append('null')
          
        if hasattr(metadata.streaminfo,'duration'):
          g=str(metadata.streaminfo.duration)
          #print(g)
          l2.append(g)
          #print(l2)
        else:
          l2.append('null')
          
        # l1.append('duration')
        print(l1)
        print(l2)
        dic=dict(zip(l1,l2))
        dic
        with open('d.json', 'a+') as outfile:
            #json.dump(dic, outfile)
            outfile.write(json.dumps(dic))
            outfile.write(",")
        l2=[]
        it=[]



```
